```python
DD_90_10_ewma = pd.read_csv('cache/DD_ewma.csv', index_col=0, parse_dates=True)
DD_garch = pd.read_csv('cache/DD_garch.csv', index_col=0, parse_dates=True)
CRA = pd.read_csv('cache/CRA.csv', index_col=0, parse_dates=True)

fig, ax = plt.subplots()
(1 + DD_90_10_ewma).cumprod().plot(label='DD_90_10_ewma', ax=ax)
(1 + DD_garch).cumprod().plot(label='DD_garch', ax=ax)
(1 + CRA).cumprod().plot(label='CRA', ax=ax)
plt.legend(['DD90/10 (EWMA)', 'DD90/10 (GARCH)', 'CRA']);
```
